## 첫 번째 머신러닝 프로그램

```python
#도미와 빙어 데이터를 하나로 합침 
length = bream_length+smelt_length
weight = bream_weight+smelt_weight
```

```python
fish_data = [[l, w] for l, w in zip(length, weight)]
#리스트로 만들어줌 (2차원 배열)

print(fish_data)
```

```python
fish_target = [1]*35 + [0]*14
print(fish_target)
```

```python
# k- 최근접 이웃 알고리즘(KNN) 

from sklearn.neighbors import KNeighborsClassifier
```

```python
# 기본 이웃 의 갯수 : 5개 (기본값) Hyper parimeter  
# n_neighbors = 5
kn = KNeighborsClassifier()
```

```python
kn.fit(fish_data, fish_target) # 훈련 
```

```python
kn.score(fish_data, fish_target) # 측정 
```

### k-최근접 이웃 알고리즘

```python
#predict_height = 30
#predict_weight = 600

predict_height = 10
predict_weight = 10


plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)

plt.scatter(predict_height, predict_weight, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
predict_y = kn.predict([[predict_height, predict_weight]])

print(predict_y) # 빙어로 나옴 
```

### 확인 문제

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
    # 최근접 이웃 개수 설정
    kn.n_neighbors = n
    # 점수 계산
    score = kn.score(fish_data, fish_target)
    print("neighbrs :", n , "Score :", score)
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    if score < 0.7:
        print(n, score)
        break
```

```python
kn17 = KNeighborsClassifier(n_neighbors=28) #이웃의 숫자를 17로 
```
